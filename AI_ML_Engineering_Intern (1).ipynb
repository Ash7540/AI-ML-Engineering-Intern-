```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
base_site = "https://finance.yahoo.com/quote/GOEV/"
```

```python
response = requests.get(base_site)
response.status_code
```

```python
html = response.content
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
with open('CanooGoev.html', 'wb') as file:
    file.write(soup.prettify('utf-8'))
```

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
# Exporting the HTML to a file
with open('CanooGoev_LXML_Parser.html', 'wb') as file:
    file.write(soup.prettify('utf-8'))
```

```python
sections = soup.find_all("section", {"class": "Bxz(bb) Mb(25px) smartphone_Px(20px)"})
sections
```

```python
sections[0].find("tbody")
```

```python
headings = [section.find("tbody") for section in sections]
headings
```
